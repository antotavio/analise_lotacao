```python
tb_cargo = pd.read_csv('https://raw.githubusercontent.com/antotavio/analise_lotacao/main/cargo.csv')
tb_departamento = pd.read_csv('https://raw.githubusercontent.com/antotavio/analise_lotacao/main/departamento.csv')
tb_funcionario = pd.read_csv('https://raw.githubusercontent.com/antotavio/analise_lotacao/main/funcionario.csv')
tb_lotacao = pd.read_csv('https://raw.githubusercontent.com/antotavio/analise_lotacao/main/lotacao.csv')
```

### Problema de Negócio

```python
query = """
select
fc.nome as nm_funcionario,
cg.nome as nm_cargo,
fc.pretensao_salarial,
lt.salario
from tb_lotacao as lt
inner join tb_funcionario as fc on fc.id = lt.id_funcionario
inner join tb_cargo as cg on cg.id = lt.id_cargo
"""

ps.sqldf(query, locals())
```

### Listar os funcionarios(nome e salário) que não sejam PCD e que possuam salário maior que R$ 4.000,00

```python
print('Tabela Lotação')
display(tb_lotacao.head(2))
print('Tabela Funcionário')
display(tb_funcionario.head(2))
```

```python
query = """
select
fc.nome,
fc.pcd,
lt.salario
from tb_lotacao as lt
inner join tb_funcionario as fc on fc.id = lt.id_funcionario
where fc.pcd = 0 and lt.salario > 4000
order by lt.salario desc
"""

ps.sqldf(query, locals())
```

### LIste os cargos e suas respectivas médias salariais maiores que R$ 5.000,00

```python
query = """
select
cg.nome,
avg(lt.salario) as média_salarial
from tb_lotacao as lt
inner join tb_cargo as cg on cg.id = lt.id_cargo
group by
cg.nome
having avg(lt.salario) > 5000
"""

ps.sqldf(query, locals())

```
